# Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of given company between two specific dates.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/national-stock-exchange-time-series/infy_stock.csv')
df
```

```python
df = df.set_index(df['Date'])
df = df['2015-01-01':'2015-01-30']

up = df[df['Close'] >= df['Open']]
down = df[df['Open'] > df['Close']]

plt.figure(figsize=(10,5))


plt.bar(x=up.index,height=up.Close-up.Open,bottom=up['Open'],width=0.3,color='green')
plt.bar(x=up.index,height=up.High-up.Low,bottom=up['Low'],width=0.05,color='green')

plt.bar(x=down.index,height=down.Close-down.Open,bottom=down['Close'],width=0.3,color='red')
plt.bar(x=down.index,height=down.High-down.Low,bottom=down['Low'],width=0.05,color='red')
plt.xticks(rotation=45, ha='right')
plt.title("Candlestick Chart",fontsize=22)
plt.xlabel("Dates",fontsize=18)
plt.ylabel("Prices",fontsize=18)
plt.rcParams['figure.facecolor'] = 'grey'
plt.rcParams['axes.facecolor'] = 'black'

plt.show()
```
