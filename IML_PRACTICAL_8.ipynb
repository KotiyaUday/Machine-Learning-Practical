# Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/kaggle/input/infosys-stock-price-till-april2023/infosys1.csv')
df
```

```python
df = df.sort_values(by=['Date'])
df = df.set_index(df['Date'])
Data = df['2019-01-02':'2019-07-31':10]
Dates = Data['Date']
Opening = Data['Open']
Closing = Data['Close']
plt.figure(figsize=(20,7))
plt.title("Stock Market")
plt.plot(Dates,Opening,label='Open',marker='o')
plt.plot(Dates,Closing,label='Close',marker='o')
plt.xlabel('Dates')
plt.ylabel('Price')
plt.grid()
plt.legend()
plt.show()
```
