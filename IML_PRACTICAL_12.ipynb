# Write a Python program using Scikit-learn to print the keys, number of rows-columns, feature names and the description of the given data.

```python
from sklearn import datasets
```

```python
data = datasets.load_iris()
```

```python
print("Keys of the dataset:\n", data.keys())
print("\nNumber of rows and columns (shape):\n", data.data.shape)
print("\nFeature names:\n", data.feature_names)
```

```python
print("\nDescription of the dataset:\n", data.DESCR)
```
