# **Write a NumPy program to implement following operation**

```python
import numpy as np
```

**1. To find the maximum and minimum value of a given flattened array**

```python
arr = np.random.randint(1,100,(3,3))
print("The original array is : \n",arr)
arr = arr.flatten()
print("The flattened array is : \n",arr)
print("The maximum value in array is : \n",np.max(arr))
print("The minimum value in array is : \n",np.min(arr))
```

**2. To compute the mean, standard deviation, and variance of a given
array along the second axis**

```python
array = np.random.randint(1,100,(3,3))
print("The array is : \n",array)
print("The mean of the array along the second axis is : \n",np.mean(array,axis=1))
print("The standard deviation of the array along the second axis is : \n",np.std(array,axis=1))
print("The variance of the array along the second axis is : \n",np.var(array,axis=1))
```
