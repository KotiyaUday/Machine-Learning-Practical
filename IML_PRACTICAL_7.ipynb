# Write a Pandas program to implement following operation

```python
import numpy as np
import pandas as pd
```

**1. To create a dataframe from a dictionary and display it**

```python
dics = {'Enrollment':[60,61,64,71,74,76],'Name':['Uday','Asfak','Rudra','Dharmit','Shivam','Shafin'],'SPI':[9.62,8.07,7.97,8.65,8.90,8.87]}
print("Dictionary\n",dics)
print()
df = pd.DataFrame(dics)
print("DataFrame\n",df)
```

**2. To sort the DataFrame first by 'name' in ascending order**

```python
print("The Sorted DataFrame\n",df.sort_values(by=['Name']))
```

**3. To delete the one specific column from the DataFrame**

```python
df['Gender'] = 'Male'
print("Adding Column to DataFrame\n",df)
del df['Gender']
print("Deleting Column From DataFrame\n",df)
```

**4. To write a DataFrame to CSV file using tab separator**

```python
df.to_csv('/kaggle/working/Student.csv',sep="\t",index=False)
print("Successfully Writing into the Student.csv by tab separator")
new_df = pd.read_csv('/kaggle/working/Student.csv')
print("\n\nThe Data of Student.csv")
new_df
```
