```python
import numpy as np
import pandas as pd
```

**1. To convert a NumPy array to a Pandas series**

```python
arr = np.random.randint(1,11,size=(10))
ser = pd.Series(arr)
print(ser)
```

**2. To convert the first column of a DataFrame as a Series**

```python
df = pd.DataFrame({'Enrollment':[60,61,64,71,74,76],'Name':['Uday','Asfak','Rudra','Dharmit','Shivam','Shafin'],'SPI':[9.62,8.07,7.97,8.65,8.90,8.87]})
print("DataFrame\n",df)
series = pd.Series(df['Enrollment'])
print("Series\n",series)
```

**3. To create the mean and standard deviation of the data of a given Series**

```python
print("Series\n",ser)
print()
print("The Mean of series : ",np.mean(ser))
print()
print("The Standard Derivation of series : ",np.std(ser))
```
