```python
import numpy as np
```

**1. To split an array of 14 elements into 3 arrays, each with 2, 4, and 8
elements in the original order**

```python
arr = np.arange(1,15)
print("The main array is : ")
print(arr)
split1, split2, split3 = np.split(arr,[2,6])
print("The splited array 1 : ")
print(split1)
print("The splited array 2 : ")
print(split2)
print("The splited array 3 : ")
print(split3)
```

**2. To stack arrays horizontally (column wise)**

```python
arr1 = np.arange(1,7).reshape(2,3)
print("The first array is : \n",arr1)
arr2 = np.arange(7,13).reshape(2,3)
print("The second array is : \n",arr2)
arr3 = np.hstack([arr1,arr2])
print("The stacked array is : \n",arr3)
```
