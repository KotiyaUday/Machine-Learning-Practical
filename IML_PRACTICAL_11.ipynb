# Write a Pandas program to filter all columns where all entries present, check which rows and columns has a NaN and finally drop rows with any NaNs from the given dataset.


```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/titanic-dataset/Titanic-Dataset.csv')
df
```

### All columns where all entries present

```python
not_null = df.dropna(axis=1)
not_null
```

### Rows and Columns has contains NaN

```python
nan_rows = df[df.isna().any(axis=1)]
print("The Rows with NaN values : ")
nan_rows
```

```python
nan_column = df.columns[df.isna().any()]
print("The Columns with NaN values : ")
nan_column
```

### Drop rows with NaNs values

```python
df.dropna(inplace=True)
df
```
