# **Write a NumPy program to implement following operation**

```python
import numpy as np
```

**1. To add, subtract, multiply, divide arguments element-wise**

```python
arr = np.random.randint(100,109,(3,3))
print("The array is : \n",arr)
print(f"array + {2} : \n{arr+2}")
print(f"array - {2} : \n{arr-2}")
print(f"array * {2} : \n{arr*2}")
div = arr / 7
print(f"array / {7} : \n{div}")
```

**2.  To round elements of the array to the nearest integer**

```python
rou = np.round(div)
print("The without rounded array is : \n",div)
print("The rounded array is : \n",rou)
```

**3. To calculate mean across dimension, in a 2D numpy array**

```python
array2D = np.arange(1,10).reshape(3,3)
print("The 2D array is : \n",array2D)
print("The mean operation of full array : \n",np.mean(array2D))
print("The mean operation of array row by row (axis = 0) : \n",np.mean(array2D,axis=0))
print("The mean operation of array column by column (axis = 1) : \n",np.mean(array2D,axis=1))
```

**4. To calculate the difference between neighboring elements, elementwise of a given array**

```python
array1D = np.random.randint(100,size = 6)
print("The array is : \n",array1D)
print("The difference between neighboring elements : \n",np.ediff1d(array1D))
```
